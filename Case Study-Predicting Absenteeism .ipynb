```python
# Encoding the Reason for Absence column
reasons=pd.get_dummies(df['Reason for Absence'])
reasons.head()
```

```python
reasons['check']=reasons.sum(axis=1)
reasons.head()
```

```python
reasons['check'].sum(axis=0)
```

```python
len(reasons)
```

```python
reasons['check'].unique()
```

```python
reasons.drop(columns='check',inplace=True)
reasons.head()
```

```python
reasons=pd.get_dummies(df['Reason for Absence'],drop_first=True)
reasons.head()
```

```python
# Grouping the reasons columns
reason_class1=reasons.loc[:,1:14].max(axis=1)
reason_class2=reasons.loc[:,15:17].max(axis=1)
reason_class3=reasons.loc[:,18:21].max(axis=1)
reason_class4=reasons.loc[:,22:].max(axis=1)
```

```python
reason_class1
```

```python
df=pd.concat([reason_class1,reason_class2,reason_class3,reason_class4,df],axis=1)
df.head()
```

```python
df.drop(columns='Reason for Absence',inplace=True)
df.head()
```

```python
df=df.rename(columns={0:'Reason_1',1:'Reason_2',2:'Reason_3',3:'Reason_4'})
df.head()
```

```python
# creating a checkpoint
df_reasons_processed=df.copy()
```

```python
# Let's process Date column
type(df_reasons_processed['Date'][0])
```

```python
df_reasons_processed.info()
```

```python
# let's convert these to datetime object
df_reasons_processed['Date']=pd.to_datetime(df['Date'],format='%d/%m/%Y')
```

```python
df_reasons_processed.head(2)
```

```python
type(df_reasons_processed['Date'][0])
```

```python
df_reasons_processed.info()
```

```python
# Extracting the Month Value
list_months=[]
for i in range(len(df_reasons_processed)):
    list_months.append(df_reasons_processed['Date'][i].month)
```

```python
len(list_months)
```

```python
df_reasons_processed['Month Value']=list_months
```

```python
df_reasons_processed.head(2)
```

```python
# Extracting the Day of the Week
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reasons_processed['Day of the Week']=df_reasons_processed['Date'].apply(date_to_weekday)
```

```python
df_reasons_processed.head(2)
```

```python
df_reasons_processed.drop(columns='Date',inplace=True)
```

```python
df_reasons_processed.head(2)
```

```python
df_reasons_processed.columns
```

```python
# Reordering the columns
column_names_reordered=['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Month Value','Day of the Week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']
df_reasons_processed=df_reasons_processed[column_names_reordered]
```

```python
df_reasons_processed.head(2)
```

```python
# Creating another checkpoint
df_reasons_date_processed=df_reasons_processed.copy()
```

```python
df_reasons_date_processed['Education'].unique()
```

```python
df_reasons_date_processed['Education'].value_counts()
```

```python
# Thus we've to apply feature scaling to all the columns except the Reason's & Education Column. Now since Education Column is
# in the middle of a df,so as to apply scaling conveniently we'll extract the education values in a variable,then drop it 
# from our df & finally insert the extracted values of eduction at our desired location in the df.
# insert 
edu=df['Education']
edu
```

```python
#splitting the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train[:,5:]=sc.fit_transform(x_train[:,5:])
x_test[:,5:]=sc.transform(x_test[:,5:])
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
x_train
```

```python
x_test
```

```python
#training the model
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression(random_state=0)
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
#calculating the accuracy
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
# Thus with our training mechanism,we got 77% accuracy which is also confirmed with the K-Fold and this is better than that of
# Udemy's Accuracy which was only 73%
```

```python
# Applying K-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies=cross_val_score(estimator=reg,X=x_train,y=y_train,cv=10)
print('Accuracy is: {:.2f} %'.format(accuracies.mean()*100))
print('StDev is: {:.2f} %'.format(accuracies.std()*100))
```

```python
# Applying Grid Search
from sklearn.model_selection import GridSearchCV
parameters=[{'C':[0.25,0.5,0.75,1.0],'penalty':['l2'],'solver':['newton-cg','lbfgs','sag'],'max_iter':[50,75,100,125,150],'multi_class':['auto','ovr','multinomial']},
            {'C':[0.25,0.5,0.75,1.0],'multi_class':['multinomial'],'solver':['newton-cg','lbfgs','sag','saga'],'max_iter':[50,75,100,125,150]}]
grid_search=GridSearchCV(estimator=reg,param_grid=parameters,scoring='accuracy',cv=10,n_jobs=-1)
grid_search.fit(x_train,y_train)
best_accuracy=grid_search.best_score_
best_parameters=grid_search.best_params_
print('Best Accuracy is: {:.2f} %'.format(best_accuracy*100))
print('Best Parameters are:',best_parameters)
```
